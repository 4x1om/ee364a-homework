```python
import numpy as np
import cvxpy as cp
```

```python
A = np.array([[1, 2, 0, 1], [0, 0, 3, 1], [0, 3, 1, 1], [2, 1, 2, 5], [1, 0, 3, 2]])
b = np.ones(5) * 100
p = np.array([3, 2, 7, 6])
d = np.array([2, 1, 4, 2])
q = np.array([4, 10, 5, 10])
```

**LP formulation**


```python
t, x = cp.Variable(4), cp.Variable(4)
obj = cp.Maximize((p - d).T @ t + d.T @ x)
cons = [0 <= t, t <= q, x >= t, A @ x <= b]

prob = cp.Problem(obj, cons)
prob.solve()

(prob.status, prob.value, t.value, x.value)
```

```python
tstar, xstar = t.value, x.value
[round(d[i] * xstar[i] + (p - d)[i] * tstar[i], 1) for i in range(4)]
[(d[i] * xstar[i] + (p - d)[i] * tstar[i]) / xstar[i] for i in range(4)]
```

**Original formulation**


```python
x = cp.Variable(4)
# this is convex + concave = not recognized
# rev = p.T @ cp.minimum(x, q) + d.T @ cp.maximum(x - q, 0)
rev = d.T @ x + (p - d).T @ cp.minimum(x, q)
obj = cp.Maximize(rev)
cons = [A @ x <= b, x >= 0]

prob = cp.Problem(obj, cons)
prob.solve()
(prob.status, prob.value, x.value)
```
